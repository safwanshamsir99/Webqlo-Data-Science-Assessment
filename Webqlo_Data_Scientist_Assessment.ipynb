## Data Cleaning

````python
df['video_views'].fillna(0, inplace=True)
df['shares'].fillna("{'count': 0}", inplace=True)
````

````python
df.isna().sum()
````

````python
df['total_count_reactions'] = df['reactions'].apply(lambda x: eval(x)['summary']['total_count'])
df.head(3)
````

````python
import json
df['shares_count'] = df['shares'].apply(lambda x: eval(x)['count'])
df.head(3)
````

````python
df['comments_total_count'] = df['interactions'] - df['video_views'] - df['shares_count'] - df['total_count_reactions']
df.head(3)
````

## Total interactions



```
"total_interactions":{
  "all":10000,
  "video_views":6913,
  "non_video_interactions":3087,
  "shares":1000,
  "reactions":2000,
  "comments":87
}
```
Formula:
- Interactions = video_views + shares + comments + reactions
- Non_video_interactions = shares + comments + reactions


````python
df['non_video_interactions'] = df['shares_count'] + df['comments_total_count'] + df['total_count_reactions']
df.head(3)
````

````python
video_views = df['video_views'].sum()
non_video_interactions = df['non_video_interactions'].sum()
shares = df['shares_count'].sum()
reactions = df['total_count_reactions'].sum()
comments = df['comments_total_count'].sum()
all = video_views + non_video_interactions

total_interactions = {
    "all": int(all),
    "video_views": int(video_views),
    "non_video_interactions": int(non_video_interactions),
    "shares": int(shares),
    "reactions": int(reactions),
    "comments": int(comments)
}

total_interactions
````

## Total post


```
"total_post":{
  "all":500,
  "photo":250,
  "video":240,
  "link":10
}
```
Information:
- total_posts are divided into post type.

````python
photo = df[df['type'] == 'photo'].shape[0]
video = df[df['type'] == 'video'].shape[0]
link = df[df['type'] == 'link'].shape[0]
all_total_post = photo + video + link

total_post = {
    "all": int(all_total_post),
    "photo": int(photo),
    "video": int(video),
    "link": int(link)
}

total_post
````

## Interactions by type

```
"interactions_by_type":[
  {
      "id":"photo",
      "name":"Photo",
      "count":10,
      "interactions":100
  },
  {
      "id":"video",
      "name":"Video",
      "count":10,
      "interactions":100
  },
  {
      "id":"others",
      "name":"Others",
      "count":10,
      "interactions":100
  }
]
```
Informations:
- Interactions are calculated as:
 1. photo
 2. video
 3. others.

 Those that are neither photo nor video will be considered as others.

````python
df['type'].unique()
````

````python
def classify_interactions(unique_val:str)->str:
    if unique_val == 'photo':
        return 'photo'
    elif unique_val == 'video':
        return 'video'
    else:
        return 'others'

df['interactions_type'] = df['type'].apply(classify_interactions)
df.head(3)
````

````python
df['interactions_type'].unique()
````

````python
photo_interactions = df.loc[df['interactions_type'] == 'photo', 'interactions'].sum()
video_interactions = df.loc[df['interactions_type'] == 'video', 'interactions'].sum()
others_interactions = df.loc[df['interactions_type'] == 'others', 'interactions'].sum()
others = df[df['interactions_type'] == 'others'].shape[0]

photo_type = {
    "id":"photo",
    "name":"Photo",
    "count": int(photo),
    "interactions": int(photo_interactions)
}

video_type = {
    "id":"video",
    "name":"Video",
    "count": int(video),
    "interactions": int(video_interactions)
}

others_type = {
    "id":"other",
    "name":"Others",
    "count": int(others),
    "interactions": int(others_interactions)
}

interactions_by_type = [photo_type, video_type, others_type]
interactions_by_type
````
